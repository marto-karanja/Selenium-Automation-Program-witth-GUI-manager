```python
import os
import time
import json
# import webdriver 
from selenium.webdriver.chrome.options import Options 
from selenium import webdriver  
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementNotVisibleException 


with open('config.json', 'r') as f:
    config = json.load(f)
    
    
user_details= {}
user_details['email'] = config['user_details']['email']
user_details['password'] = config['user_details']['password']
user_details

CHROME_DRIVER = 'c:\\Users\\Admin\\Documents\\projects\\bots\\writedom\\bot\\bot\\chromedriver\\chromedriver.exe'



# create webdriver object  
chrome_options = Options()
chrome_options.add_argument("--incognito")
chrome_options.add_argument("--start-maximized")
driver = webdriver.Chrome(options=chrome_options, executable_path=CHROME_DRIVER)

url = "https://essayshark.com/"  
driver.get(url)
#log in
## click login apply button
### wait for form to load before clicking login button
try:
    # get element  after explicitly waiting for 10 seconds
    WebDriverWait(driver, 120).until( 
        EC.element_to_be_clickable((By.ID, "id_esauth_myaccount_login_link")))
except (ElementNotVisibleException, NoSuchElementException) as e:
    traceback.format_exc(e)
login_form = driver.find_element_by_id("id_esauth_myaccount_login_link")
login_form.click()
### wait for form to load before entering details
try:
    # get element  after explicitly waiting for 10 seconds
    WebDriverWait(driver, 120).until( 
        EC.element_to_be_clickable((By.ID, "id_esauth_login_button")))
except (ElementNotVisibleException, NoSuchElementException) as e:
    traceback.format_exc(e)
username = driver.find_element_by_id('id_esauth_login_field')
password = driver.find_element_by_id('id_esauth_pwd_field')
username.send_keys(user_details['email'])
password.send_keys(user_details['password'])
signin_button  = driver.find_element_by_id("id_esauth_login_button")
signin_button.click()
# fetch orders
## wait for page to be ready before fetching orders
try:
    # get element  after explicitly waiting for 10 seconds
    WebDriverWait(driver, 120).until( 
        EC.visibility_of_any_elements_located((By.CLASS_NAME, "order_number")))
except (ElementNotVisibleException, NoSuchElementException) as e:
    traceback.format_exc(e)
order_url_element = driver.find_elements_by_css_selector('td.order_number>a')
order_urls = set()
for element in order_url_element:
    order_urls.add(element.get_attribute("href"))
# print urls in table    
print(order_urls)
# apply to order
## create a set of applied orders
applied_orders = set()
###get difference bteween applied orders and new orders
processing_orders = applied_orders.difference(order_urls)
### loop through available orders url
for order in processing_orders:
    #### fetch url
    driver.get(order)
    time.sleep(145)
    ##process url
    #### store the applied urls in the applied orders set
    applied_orders.add(order)
# wait and fetch new order

print(applied_orders)

time.sleep(25)
#log out
"""
signout_button = driver.find_element_by_class_name("auth_user_menu_authlogout.html")
signout_button.click()"""
```
